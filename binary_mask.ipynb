```python
import torch
import cv2
from google.colab.patches import cv2_imshow

def generate_mask_rgb(I, B, threshold=30):
    # Compute the absolute difference between the degraded image and clean image
    diff = torch.abs(I - B)

    print(diff)
    # Convert difference to grayscale by taking the mean across color channels
    diff_gray = torch.mean(diff, dim=2, keepdim=True)

    # Apply threshold to generate the mask
    mask = (diff_gray >= threshold).float()
    print(mask.shape)
    return mask

I = torch.from_numpy(cv2.imread('right_1535201264389466.png')*1.0)
B = torch.from_numpy(cv2.imread('left_1535201264389466.png')*1.0)
print(I.shape)
#print(I,B)
mask = generate_mask_rgb(I,B)
cv2_imshow(mask.numpy()*256)
print(mask)
```
